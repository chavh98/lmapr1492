```python
from pymatgen.io.cif import CifParser
parser = CifParser("HoCuO2_mp-754472_symmetrized.cif")
from pymatgen.ext.matproj import MPRester
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
with MPRester("LqPM0vGDumeSiU5h") as m:

#Structure for material id
    structure = m.get_structure_by_material_id("mp-754472")

#coordonnées des atomes
atome1_Ho0=[0, 0, 0.5]
atome2_Cu1=[0, 0, 0]
atome3_O2=[0, 0, 0.11]

#opérations sur les atomes   
operation1=SymmOp.from_xyz_string('-y,-x,z')
operation2=SymmOp.from_xyz_string('-x,-y,-z')
operation3=SymmOp.from_xyz_string('y,-x+y,-z')

#résultats des opérations de symétrie
atomesym1=operation1.operate(atome1_Ho0)
atomesym2=operation2.operate(atome2_Cu1)
atomesym3=operation3.operate(atome3_O2)

print("Opération 1 de symétrie :", operation1, "Atome au départ :", atome1_Ho0, "Atome après l'opération de symmétrie :", atomesym1)
print("Opération 2 de symétrie :", operation2, "Atome au départ :", atome2_Cu1, "Atome après l'opération de symmétrie :", atomesym2)
print("Opération 3 de symétrie :", operation3, "Atome au départ :", atome3_O2, "Atome après l'opération de symmétrie :", atomesym3)

#images pour illustrer
image1=mpimg.imread('MirrorPlane.PNG') #operation1
image2=mpimg.imread('Ci.PNG') #operation2
image3=mpimg.imread('3-barAxis.PNG') #operation3

plt.title('(-y, -x, z) sur Ho0')
plt.imshow(image1)

plt.figure()
plt.title('(-x, -y, -z) sur Cu1')
plt.imshow(image2)

plt.figure()
plt.title('(y, -x+y, -z) sur O2')
plt.imshow(image3)
    
```

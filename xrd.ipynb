```python
from pymatgen.io.cif import CifParser
parser = CifParser("HoCuO2_mp-754472_symmetrized.cif")
from pymatgen.ext.matproj import MPRester
with MPRester("LqPM0vGDumeSiU5h") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-754472")

from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt

xrdCalculator=XRDCalculator(wavelength='CuKa',symprec=0.1,debye_waller_factors=None)
spectre=xrdCalculator.get_pattern(structure)

#les 3 premiers pics du diffractogramme 
pic1=(spectre.hkls)[0][0]["hkl"]
pic2=(spectre.hkls)[1][0]["hkl"]
pic3=(spectre.hkls)[2][0]["hkl"]

print("Premier pic :",pic1)
print("Deuxième pic :",pic2)
print("Troisième pic :",pic3)

xrdCalculator.show_plot(structure)

```

# Symétrie
Dans ce dossier, trois éléments de symétrie du cristal seront étudiés. Ils seront chacun appliqué à un atome différent.

```python
from pymatgen.io.cif import CifParser
parser = CifParser("HoCuO2_mp-754472_symmetrized.cif")
from pymatgen.ext.matproj import MPRester
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt
with MPRester("LqPM0vGDumeSiU5h") as m:

#Structure for material id
    structure = m.get_structure_by_material_id("mp-754472")

#coordonnées des atomes
atome1_Ho0=[0, 0, 0.5]
atome2_Cu1=[0, 0, 0]
atome3_O2=[0, 0, 0.11]

#opérations sur les atomes   
operation1=SymmOp.from_xyz_string('-y,-x,z')
operation2=SymmOp.from_xyz_string('-x,-y,-z')
operation3=SymmOp.from_xyz_string('y,-x+y,-z')

#résultats des opérations de symétrie
atomesym1=operation1.operate(atome1_Ho0)
atomesym2=operation2.operate(atome2_Cu1)
atomesym3=operation3.operate(atome3_O2)

print("Opération 1 de symétrie :", operation1, "Atome au départ :", atome1_Ho0, "Atome après l'opération de symmétrie :", atomesym1)
print("Opération 2 de symétrie :", operation2, "Atome au départ :", atome2_Cu1, "Atome après l'opération de symmétrie :", atomesym2)
print("Opération 3 de symétrie :", operation3, "Atome au départ :", atome3_O2, "Atome après l'opération de symmétrie :", atomesym3)
```

### Première opération de symétrie
La première opération de symétrie est un plan miroir. Il s'agit d'appliquer la transformation suivante 
$\begin{pmatrix}
    -y & -x & z 
\end{pmatrix}$
à un atome. La matrice associée à l'opération est :
$\begin{pmatrix}
    0 & -1 & 0 \\ -1 & 0 & 0 \\ 0 & 0 & 1 
\end{pmatrix}$

L'atome Ho situé en 
$\begin{pmatrix}
    0 & 0 & 0.5 
\end{pmatrix}$
qui subit cet élément de symétrie reste en 
$\begin{pmatrix}
    0 & 0 & 0.5 
\end{pmatrix}$. La situation est représentée sur la figure suivante où le plan miroir est représenté en vert. 

```python
image1=mpimg.imread('MirrorPlane.PNG') #operation1
fig, ax = plt.subplots(figsize=(9,9))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.title('(-y, -x, z) sur Ho0')
plt.imshow(image1)
plt.show()
```

### Deuxième opération de symétrie
La deuxième opération de symétrie est une inversion. Il s'agit simplement d'inverser le signe des composantes par la transformation suivante 
$\begin{pmatrix}
    -x & -y & -z 
\end{pmatrix}$
à un atome. La matrice associée à l'opération est :
$\begin{pmatrix}
    -1 & 0 & 0 \\ 0 & -1 & 0 \\ 0 & 0 & -1 
\end{pmatrix}$

L'atome Cu situé en 
$\begin{pmatrix}
    0 & 0 & 0 
\end{pmatrix}$
qui subit cet élément de symétrie reste en 
$\begin{pmatrix}
    0 & 0 & 0. 
\end{pmatrix}$. La situation est représentée sur la figure suivante.

```python
image2=mpimg.imread('Ci.PNG') #operation2
fig, ax = plt.subplots(figsize=(9,9))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.title('(-x, -y, -z) sur Cu1')
plt.imshow(image2)
plt.show()
```

### Troisième opération de symétrie
La troisième opération de symétrie est une roto-inversion d'ordre 3. Plus précisément, c'est une rotation d'ordre 3 suivi d'une inversion de signe des composantes. La transformation envoyée à un atome est 
$\begin{pmatrix}
    y & -x+y & -z 
\end{pmatrix}$, et la matrice associée à l'opération est :
$\begin{pmatrix}
    0 & 1 & 0 \\ -1 & 1 & 0 \\ 0 & 0 & -1 
\end{pmatrix}$.

L'atome O situé en 
$\begin{pmatrix}
    0 & 0 & 0.11 
\end{pmatrix}$
qui subit cet élément de symétrie arrive en 
$\begin{pmatrix}
    0 & 0 & -0.11 
\end{pmatrix}$. La situation est représentée sur la figure suivante.

```python
image3=mpimg.imread('3-barAxis.PNG') #operation3
fig, ax = plt.subplots(figsize=(9,9))
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)
plt.title('(y, -x+y, -z) sur O2')
plt.imshow(image3)
plt.show()
```
